```python
# Import the necessary libraries
import requests

# Define the URL of the book on Project Gutenberg
url = "https://www.gutenberg.org/ebooks/236.txt.utf-8"

# Send a GET request to the URL
response = requests.get(url)

# Save the content of the response to a text file
with open("Jungle_Book.txt", "w") as file:
    file.write(response.text)
```

```python
# Install NLTK
!pip install nltk

# Import necessary libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.stem import PorterStemmer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download NLTK resources (stopwords and tokenizer models)
nltk.download("stopwords")
nltk.download("punkt")
```

```python
# Read the downloaded text file
with open("Jungle_Book.txt", "r") as file:
    text = file.read()

# Tokenize the text into words
words = word_tokenize(text)

# Remove stopwords
stop_words = set(stopwords.words("english"))
filtered_words = [word for word in words if word.lower() not in stop_words]

# Initialize the Porter Stemmer
stemmer = PorterStemmer()

# Perform stemming on the filtered words
stemmed_words = [stemmer.stem(word) for word in filtered_words]
```

```python
print(stemmed_words)
```

```python
print(text)
```

```python
# Calculate word frequencies
word_freq = FreqDist(stemmed_words)

# Get the 50 most common words
most_common_words = word_freq.most_common(50)

# Plot a histogram of word frequencies
plt.figure(figsize=(12, 6))
word_freq.plot(50, cumulative=False)
plt.title("Top 50 Most Frequent Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
plt.show()
```

```python
# Convert the list of stemmed words back to a string
filtered_text = " ".join(filtered_words)

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(filtered_text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud")
plt.show()
```

```python
# Get a list of word-frequency pairs sorted by frequency
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
```

```python
print(sorted_word_freq)
```

```python
# Create a list of word ranks
word_ranks = list(range(1, len(sorted_word_freq) + 1))
```

```python
print(word_ranks)
```

```python
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Convert frequencies and ranks to NumPy arrays
log_freq = np.log([pair[1] for pair in sorted_word_freq])
log_rank = np.log(word_ranks)

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(log_rank, log_freq)

# Calculate α and C
alpha = -slope
corpus_constant = np.exp(intercept)
# brainstorm a bit on the valid values for alpha and C!!!

print("Exponent (α):", alpha)
print("Corpus Constant (C):", corpus_constant)
```

```python
# Plot Zipf's Law
plt.figure(figsize=(10, 5))
plt.scatter(log_rank, log_freq, s=5, alpha=0.5)
plt.plot(log_rank, intercept + slope * log_rank, color='red', linestyle='--')
plt.title("Zipf's Law")
plt.xlabel("log(rank)")
plt.ylabel("log(frequency)")
plt.show()
```

```python
# Plot Zipf's Law (2)
plt.figure(figsize=(10, 5))
plt.scatter(word_ranks, sorted_word_freq, s=5, alpha=0.5)
plt.plot(word_ranks, intercept + slope * word_ranks, color='red', linestyle='--')
plt.title("Zipf's Law (2)")
plt.xlabel("word_ranks")
plt.ylabel("word_freq")
plt.show()
```
